### Exercise 5 - parts a & b

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.style.use('ggplot')
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

house_price = pd.read_csv('Real_Estate.csv')

#Drop NO and transaction_date columns
house_price = house_price.drop(columns = ['No', 'transaction_date'], axis = 1)
house_price.head()
```

### Exercise 5 - part c

```python
#Input and target variable
X = house_price.drop(columns = 'house_price_of_unit_area', axis = 1) #Everything but this is input variables
Y = house_price['house_price_of_unit_area'] #Target variable

#Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

### Exercise 5 - part d

```python
#Store MSE results here
model_1 = list()

### Model 1 ###
X_train_1 = X_train[['house_age', 'distance_to_the_nearest_MRT_station', 'number_of_convenience_stores', 'latitude', 'longitude']]
X_test_1 = X_test[['house_age', 'distance_to_the_nearest_MRT_station', 'number_of_convenience_stores', 'latitude', 'longitude']]

lm_md_1 = LinearRegression().fit(X_train_1, Y_train)

#Predict
lm_md_pred_1 = lm_md_1.predict(X_test_1)

#Compute MSE for model 1
mse_1 = mean_squared_error(Y_test, lm_md_pred_1)
model_1.append(mse_1)
print('Model 1 MSE: ', model_1)
```

### Exercise 5 - part e

```python
#Store MSE results here
model_2 = list()

### Model 2 ###
#Got rid of distance_to_the_nearest_MRT_station for this model
X_train_2 = X_train[['house_age', 'number_of_convenience_stores', 'latitude', 'longitude']]
X_test_2 = X_test[['house_age', 'number_of_convenience_stores', 'latitude', 'longitude']]

lm_md_2 = LinearRegression().fit(X_train_2, Y_train)

#Predict
lm_md_pred_2 = lm_md_2.predict(X_test_2)

#Compute MSE for model 2
mse_2 = mean_squared_error(Y_test, lm_md_pred_2)
model_2.append(mse_2)
print('Model 2 MSE: ', model_2)
```

### Exercise 5 - part f

**I would use model 1 in part d because the MSE is lower which means that we are closer to the actual values**

### Exercise 5 - part g

```python
#Input and target variable
X = house_price.drop(columns = 'house_price_of_unit_area', axis = 1) #Everything but this is input variables
Y = house_price['house_price_of_unit_area'] #Target variable

#Store MSE results here
model_1 = list()
model_2 = list()

#100 Iterations
for i in range(0,100):
    #Split the data
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

    ### Model 1 ###
    X_train_1 = X_train[['house_age', 'distance_to_the_nearest_MRT_station', 'number_of_convenience_stores', 'latitude', 'longitude']]
    X_test_1 = X_test[['house_age', 'distance_to_the_nearest_MRT_station', 'number_of_convenience_stores', 'latitude', 'longitude']]

    lm_md_1 = LinearRegression().fit(X_train_1, Y_train)

    #Predict
    lm_md_pred_1 = lm_md_1.predict(X_test_1)

    #Compute MSE for model 1
    mse_1 = mean_squared_error(Y_test, lm_md_pred_1)
    model_1.append(mse_1)
    print('Model 1 MSE: ', model_1)


    ### Model 2 ###
    #Got rid of distance_to_the_nearest_MRT_station for this model
    X_train_2 = X_train[['house_age', 'number_of_convenience_stores', 'latitude', 'longitude']]
    X_test_2 = X_test[['house_age', 'number_of_convenience_stores', 'latitude', 'longitude']]

    lm_md_2 = LinearRegression().fit(X_train_2, Y_train)

    #Predict
    lm_md_pred_2 = lm_md_2.predict(X_test_2)

    #Compute MSE for model 2
    mse_2 = mean_squared_error(Y_test, lm_md_pred_2)
    model_2.append(mse_2)
    print('Model 2 MSE: ', model_2)

#Visual of each iteration
numb_of_runs = [i for i in range(0, 100)]
sns.lineplot(x = numb_of_runs, y = model_1, label = 'Linear Model 1')
sns.lineplot(x = numb_of_runs, y = model_2, label = 'Linear Model 2')
plt.xlabel('Number of simulation')
plt.ylabel('MSE')
```
