<h2>Exercise 6 - Use framingham datafile to answer questions below</h2>

```python
#6a -  Using the pandas library, read the csv data file and create a data-frame called heart

import pandas as pd

heart = pd.read_csv('framingham(4).csv')
heart.head()
```

```python
#6b - Remove observations with missing values.

heart = heart.dropna()
```

```python
### 6c ###

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

#Store feature importance here
importance = list()

#Repeating 100 times
for i in range(0, 100):

    #TenYearCHD is target variable and everything else is input
    X = heart.drop(columns = ['TenYearCHD'], axis = 1)
    Y = heart['TenYearCHD']

    #Split data into 80% train and 20% test
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y)

    #Random forest classifier model
    RF_md = RandomForestClassifier(n_estimators = 500).fit(X_train, Y_train)

    #Extract feature importance
    importance.append(RF_md.feature_importances_)
```

```python
# Compute the average importance of each of the variables across the 100 splits. After that, select the 
# top 5 variables (the ones with top 5 average importance) as the predictor variables.

import numpy as np
from sklearn.metrics import recall_score

importance = pd.DataFrame(importance)
importance.columns = ['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds', 
                      'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP', 
                      'diaBP', 'BMI', 'heartRate', 'glucose']

sorted = pd.DataFrame({'Importance':importance.apply(np.mean, axis = 0)})
sorted.sort_values(by = 'Importance', ascending = False)
```

```python
### 6d ###

from sklearn.ensemble import ExtraTreesClassifier

#Store recall values here
md1_results = list()
md2_results = list()
md3_results = list()
md4_results = list()
md5_results = list()
md6_results = list()

#Use top 5 variables from part c as predictor variables and TenYearCHD as target
X = heart[['sysBP', 'BMI', 'age', 'totChol', 'glucose']]
Y = heart['TenYearCHD']

#100 iterations
for i in range(0,100):

    #Split data
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y)

    
    ### RF model 500 estimators and max_depth 3 ### 
    RF_md1 = RandomForestClassifier(max_depth = 3, n_estimators = 500).fit(X_train, Y_train)
    
    #Predict
    RF_pred1 = RF_md1.predict_proba(X_test)[:,1]
    
    #Labels with 10% cutoff
    RF_label1 = np.where(RF_pred1 < .1, 0, 1)
    
    #Recall
    RF_recall1 = recall_score(Y_test, RF_label1)
    md1_results.append(RF_recall1)

    
    ### RF model 500 estimators and max_depth 5 ###
    RF_md2 = RandomForestClassifier(max_depth = 5, n_estimators = 500).fit(X_train, Y_train)
    
    #Predict
    RF_pred2 = RF_md2.predict_proba(X_test)[:,1]
    
    #Labels with 10% cutoff
    RF_label2 = np.where(RF_pred2 < .1, 0, 1)
    
    #Recall
    RF_recall2 = recall_score(Y_test, RF_label2)
    md2_results.append(RF_recall2)


    ### RF model 500 estimators and max_depth 7 ###
    RF_md3 = RandomForestClassifier(max_depth = 7, n_estimators = 500).fit(X_train, Y_train)
    
    #Predict
    RF_pred3 = RF_md3.predict_proba(X_test)[:,1]

    #Labels with 10% cutoff
    RF_label3 = np.where(RF_pred3 < .1, 0, 1)
    
    #Recall
    RF_recall3 = recall_score(Y_test, RF_label3)
    md3_results.append(RF_recall3)


    
    ### ET model 500 estimators and max_depth 3 ###
    ET_md1 = ExtraTreesClassifier(max_depth = 3, n_estimators = 500).fit(X_train, Y_train)
    
    #Predict
    ET_pred1 = ET_md1.predict_proba(X_test)[:,1]

    #Labels with 10% cutoff
    ET_label1 = np.where(ET_pred1 < .1, 0, 1)
    
    #Recall
    ET_recall1 = recall_score(Y_test, ET_label1)
    md4_results.append(ET_recall1)


    ### ET model 500 estimators and max_depth 5 ###
    ET_md2 = ExtraTreesClassifier(max_depth = 5, n_estimators = 500).fit(X_train, Y_train)
    
    #Predict
    ET_pred2 = ET_md2.predict_proba(X_test)[:, 1]

    #Labels with 10% cutoff
    ET_label2 = np.where(ET_pred2 < .1, 0, 1)
    
    #Recall
    ET_recall2 = recall_score(Y_test, ET_label2)
    md5_results.append(ET_recall2)


    ### ET model 500 estimators and max_depth 7 ###
    ET_md3 = ExtraTreesClassifier(max_depth = 7, n_estimators = 500).fit(X_train, Y_train)
    
    #Predict
    ET_pred3 = ET_md3.predict_proba(X_test)[:, 1]

    #Labels with 10% cutoff
    ET_label3 = np.where(ET_pred3 < .1, 0, 1)
    
    #Recall
    ET_recall3 = recall_score(Y_test, ET_label3)
    md6_results.append(ET_recall3)


print('Model 1 average recall: ', np.mean(md1_results))
print('Model 2 average recall: ', np.mean(md2_results))
print('Model 3 average recall: ', np.mean(md3_results))
print('Model 4 average recall: ', np.mean(md4_results))
print('Model 5 average recall: ', np.mean(md5_results))
print('Model 6 average recall: ', np.mean(md6_results))
```
